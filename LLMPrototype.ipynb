```python
from langchain_community.document_loaders import PyMuPDFLoader
from pathlib import Path
import os
import re

# Define the directory path
path = "/content/RAG_Data/"
fileNames = os.listdir(path)

# Initialize a list to hold all documents' data
all_docs = []

# Function to clean extracted text
def clean_text(text):
    # Remove extra newlines and whitespace
    text = re.sub(r'\n+', ' ', text)  # Replace multiple newlines with a single space
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space

    # Remove excessive dots (e.g., ellipsis "...")
    text = re.sub(r'\.{2,}', '.', text)  # Replace consecutive dots with a single dot

    # Optionally, remove any other unwanted characters (e.g., punctuation)
    text = re.sub(r'[^\w\s.,?!;:"\'()-]', '', text)  # Keep only common punctuation

    # Trim leading/trailing spaces
    text = text.strip()

    return text

# Loop through all the files in the directory
for fileName in fileNames:
    filePath = os.path.join(path, fileName)

    # Check if the file is a PDF
    if fileName.endswith('.pdf'):
        loader = PyMuPDFLoader(filePath)

        # Load pages asynchronously
        async def load_pages():
            pages = []
            async for page in loader.alazy_load():
                pages.append(page)
            return pages

        # Collect pages and clean them before adding to all_docs
        pages = await load_pages()
        for page in pages:
            # Clean the text of each page
            cleaned_text = clean_text(page.page_content)
            all_docs.append(cleaned_text)

# Print the number of pages loaded and a sample
print(f"Total pages loaded: {len(all_docs)}")
if all_docs:
    print("Sample page content:")
    print(all_docs)  # Print the content of the first page after cleaning
```

```python
import torch
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer

N_GPU = torch.cuda.device_count()
DEVICE = torch.device(f'cuda:{N_GPU-1}' if torch.cuda.is_available() and N_GPU > 0 else 'cpu')



model_name = "BAAI/bge-large-en-v1.5"
encoder = SentenceTransformer(model_name, device=DEVICE)


EMBEDDING_DIM = encoder.get_sentence_embedding_dimension()
MAX_SEQ_LENGTH_IN_TOKENS = encoder.get_max_seq_length()


# Load the tokenizer for the model

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Get the maximum sequence length
MAX_SEQ_LENGTH = tokenizer.model_max_length

print(f"model_name: {model_name}")
print(f"EMBEDDING_DIM: {EMBEDDING_DIM}")
print(f"MAX_SEQ_LENGTH: {MAX_SEQ_LENGTH}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
import numpy as np
from sentence_transformers import SentenceTransformer

# Assuming dataset is a list of cleaned text strings from your PDFs
#This should be the cleaned text from your PDFs

# Define chunk size and overlap
CHUNK_SIZE = 512
chunk_overlap = np.round(CHUNK_SIZE * 0.10, 0)
print(f"chunk_size: {CHUNK_SIZE}, chunk_overlap: {chunk_overlap}")

# Create an instance of the text splitter
child_splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=chunk_overlap
)

# Convert the raw text strings into Document objects
documents = [Document(page_content=text) for text in dataset]

# Split the documents into chunks
chunks = child_splitter.split_documents(documents)
print(f"{len(documents)} docs split into {len(chunks)} child documents.")

print("Started embeding vectors")
# List to store the page content
list_of_strings = [doc.page_content for doc in chunks if hasattr(doc, 'page_content')]

# Load a pre-trained Sentence-BERT model
model = encoder  # Or choose another model with higher dimensions

# Convert list of strings into embeddings (e.g., 384 or 768-dimensional vectors)
embeddings = model.encode(list_of_strings)

# Normalize embeddings (ensure they have unit length)
embeddings = np.array(embeddings)

# Normalize each embedding to unit length
embeddings = embeddings / np.linalg.norm(embeddings, axis=1)[:, np.newaxis]

# Convert embeddings to float32
converted_values = list(map(np.float32, embeddings))

# Create a dictionary with chunk content and embeddings
dict_list = []
for chunk, vector in zip(chunks, converted_values):
    # Assemble embedding vector, original text chunk, and metadata
    chunk_dict = {
        'chunk': chunk.page_content,
        'source': chunk.metadata.get('source', ""),  # Assuming 'source' exists in metadata
        'vector': vector,
    }
    dict_list.append(chunk_dict)

print("Finished")
```

```python
from pymilvus import MilvusClient
import time
mc = MilvusClient("milvus_demo.db")


# Create a collection with flexible schema and AUTOINDEX.
COLLECTION_NAME = "MilvusDocs5"
mc.create_collection(COLLECTION_NAME,
       EMBEDDING_DIM,
       consistency_level="Eventually",
       auto_id=True,
       overwrite=True)


# Insert data into the Milvus collection.
print("Start inserting entities")
start_time = time.time()
mc.insert(
   COLLECTION_NAME,
   data=dict_list,
   progress_bar=True)


end_time = time.time()
print(f"Milvus insert time for {len(dict_list)} vectors: ", end="")
print(f"{round(end_time - start_time, 2)} seconds")
```

```python
import torch
import numpy as np
import torch.nn.functional as F
from pymilvus import Collection, connections

# Connect to Milvus


# Constants
SAMPLE_QUESTION = "Što je multisport program"
COLLECTION_NAME = "MilvusDocs5"
anns_field = "vector"  # Replace 'vector' with the actual field name used for embeddings

print("Embeddings started")
# Embed the question using the same encoder
query_embeddings = torch.tensor(encoder.encode(SAMPLE_QUESTION))

# Normalize embeddings to unit length
query_embeddings = F.normalize(query_embeddings, p=2, dim=0)

# Convert the embeddings to list of lists of float
query_embeddings = query_embeddings.tolist()
query_embeddings = [[float(x) for x in query_embeddings]]  # Reshape to match Milvus requirements

print("Embeddings done")

# Define metadata fields you can filter on
OUTPUT_FIELDS = list(dict_list[0].keys())
OUTPUT_FIELDS.remove('vector')

# Define search parameters
search_params = {"metric_type": "COSINE", "params": {"nprobe": 10}}
TOP_K = 2

mc = MilvusClient("milvus_demo.db")

print("Search start")
# Run semantic vector search

results = mc.search(
    COLLECTION_NAME,
    data=query_embeddings,

    search_params=search_params,  # Correct argument name
    output_fields=OUTPUT_FIELDS,
    limit=TOP_K,
    consistency_level="Eventually"
)
# Extract relevant contexts and sources
# Corrected lines to handle results correctly
# Iterate over each list in the results
contexts = []
sources = []
for result_list in results:
    for result in result_list:
        contexts.append(result['entity'].get("chunk"))
        sources.append(result['entity'].get("source"))
print("Search finished")
# Process and display results
print("Results")
for result in results:
    print(result)

print("Context")
for context in contexts:
    print(context)
print("Sources")
for source in sources:
    print(source)
```

```python
contexts_combined = ' '.join(contexts)
# Lance Martin, LangChain, says put the best contexts at the end.
contexts_combined = ' '.join(reversed(contexts))


# Separate all the unique sources together by comma.
source_combined = ' '.join(reversed(list(dict.fromkeys(sources))))


SYSTEM_PROMPT = f"""First, check if the provided Context is relevant to
the user's question.  Second, only if the provided Context is strongly relevant, answer the question using the Context.  Otherwise, if the Context is not strongly relevant, answer the question without using the Context.
Be clear, concise, relevant.  Answer clearly, in fewer than 2 sentences.
Grounding sources: {source_combined}
Context: {contexts_combined}
User's question: {SAMPLE_QUESTION}
"""


prompts = [SYSTEM_PROMPT]
```
